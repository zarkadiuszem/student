```python
# co zwróci False
lista = []
krotka = ()
slownik = {}
zbior = set()
print(type(lista), type(krotka), type(slownik), type(zbior))
print(bool(lista), bool(krotka), bool(slownik), bool(zbior)) # puste dadzą False

lista = [1]
krotka = (1,)
slownik = {1:1}
zbior = set(lista)
print(type(lista), type(krotka), type(slownik), type(zbior))
print(bool(lista), bool(krotka), bool(slownik), bool(zbior))

print(bool(0))    # da False
print(bool(0.0))  # da False
print(bool(1))
print(bool(-1))
print(bool(0.1))

print(bool(""))   # da False
print(bool("a"))
```

```python
a = 10
b = 20
c = 30

if a < b and b > c or c > 20:   # operatory porównania najwyższy priorytet, później not, and, na końcu or
  print('o to chodziło')        # jeśli chcemy zmienić kolejność sprawdzania ()
else:
  print("coś nie tak")

print(a > b and b > c or c > 20)    # jeśli chcemy zmienić kolejność sprawdzania ()
print(a > b and (b > c or c > 20))

print(a == 10 and not b == 21) #
```

```python
# inaczej - jeśli jest dużo warunków wrzucamy je w listę
warunki = [a > b, b > c, c > 20]

if all(warunki):  # funkcja all() zwróci True jeśli wsystkie warunki są spełnione - inaczej False
  print('o to chodziło')
else:
  print("coś nie tak")
```

### Operator walrus (mors)
<h1> := <h1>

```python
x = 2
y = 3
if z := x - y > 0:  # możemy przechwycić wartość logiczną sprawdzenia (jednocześnie sprawdzamy i przypisyjemy)
  print(z)

print(z)
```

## Pętla while
dopóki spełniony jest warunek wykonuje się dany fragment kodu

```python
while 1:
  print("to jest pętla nieskończona")
```

```python
while 2 == 3:
  print("to się nie wykona")
```

```python
n = 5
e = 2
while e <= n:
  print(e)
  e += 1

print(f"e jest teraz: {e}")
```

```python
haslo = input('Podaj hasło: ')
while haslo != "poprawne hasło":
  haslo = input('Błędne hasło. Podaj poprawne hasło: ')
print("witaj w aplikacji")
```

```python
haslo = input('Podaj hasło: ')
poprawne_haslo = 'Grzyb09'
i = 1

while haslo != poprawne_haslo and i <= 2:
  haslo = input('Błędne hasło. Podaj poprawne hasło: ')
  i += 1

if haslo == poprawne_haslo:
  print("Witaj w aplikacji")
else:
  print("Próba logowania nie powiodła się")
```

```python
liczba = int(input('Podaj liczbę: '))
i = 2
while i <= 4:
  wynik = liczba ** i
  print(liczba, 'do potęgi', i, '=', wynik)
  i = i+1
```

### Instrukcja break
przerwij wykonywanie pętli

```python
for litera in 'Szczecin jest fajny':
  if litera == 'j':
    break
  print(litera, end = "") # argument end umożliwia zmianę domyślnego łamania linii w funkcji print

'kod idzie dalej'
```

```python
for litera in reversed('Szczecin jest fajny'):
  if litera == 'j':
    break
  print(litera, end = ",")
```

###Instrukcja continue
pomijamy część instrukcji

```python
liczby = [3,4,5,6,8,10]
for i in liczby:
  if i == 8:
    continue
  print(i)
```

```python
numbers = [1,2,3,4,5,6,7]
parzyste = []
nieparzyste = []
liczbaParzystych = 0
liczbaNieparzystych = 0
for i in numbers:
  if i % 2 == 0:
    parzyste.append(i)
    liczbaParzystych += 1
  else:
    nieparzyste.append(i)
    liczbaNieparzystych += 1
print(f'Mamy {liczbaParzystych} liczby parzyste: {parzyste}')
print(f'Mamy {liczbaNieparzystych} liczby nieparzyste: {nieparzyste}')
```

### Zagnieżdzone pętle
*nested loops*

```python
for i in range (1,4):     # zewnętrzna pętla
  for j in range (1,4):   # wewnętrzna pętla
    wynik = i * j
    print(i, 'x', j, '=', wynik)
  print()
```
