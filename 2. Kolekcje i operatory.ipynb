## Tuples - krotki
**typ: tup** (wym: tupls) <br>
Uporządkowana niezmienna sekwencja obiektów <br>
(2021, "arek", 20.4)<br>
zajmują mniej miejsca niż listy - szybsze - bardziej wydajne<br>
przez przypadek ich nie zmodyfikujemy<br>
ale - mniejsza funkcjonalność<br>
niezmienność oznacza też określoną długość
<h1> ( ... , ... ) </h1>

```python
#tuple
a = ('arek','b','c','s',3.3,3)
type(a)
print(a[0])

# może być też bez nawiasów
b = 1,2,3,'marek'
print(b)
#zmiana tuple - nie można zmienić elementów w tuple (is immutable) to główna różnica w porównaniu do listy
#jeśli nie chcemy czegoś zmieniac (np to będą ID studentów które nigdy się nie zmianiają) to tuple
```

```python
# tu dostajemy błąd:
# TypeError: 'tuple' object does not support item assignment
a[0]='darek'
```

```python
# przy jednoelementowych musi się pojawić przecinek - dwa sposoby
a = 'arek',
a = ('arek',)
type(a)
```

```python
#rozpakowanie krotki do zmiennych
imiona = 'arek', 'darek', 'jarek', 'marek'
a,b,c,d = imiona
a
```

```python
#zmiana krotki w listę
krotka = (1,2,3)
lista = list(krotka)
print(type(lista))
```

```python
#niemutowalne - nie zmieniamy jej tylko tworzymy nową
krotka1 = (1,2,3)
krotka2 = (4,5)
print(id(krotka1))
krotka1 = krotka1 + krotka2
print(id(krotka1))
krotka1
```

## Dictionaries - typ słownik - 5
Nieuporządkowane pary klucz - wartość <br>
Nie ma indeksów - zastępują je klucze <br>
{"imię": "arek", "wiek": 18} <br>
indeksy nie będą działały
<h1>{ ... : ... }</h1>

```python
#dictionary - nieuporządkowana kolekcja elementów ma pary klucz - wartość
dane = {'nazwisko':'gronostaj', 'płeć':'M', 'wzrost':170}
print(type(dane))
print(dane)

print(dane['nazwisko']) #wskazujemy klucz żeby uzyskać wartość
x = 'gronostaj'
print(x)


```

```python
# słowniki są mutowalne = może my dodać/skasować/zmienić element
print(dane)
dane['wiek'] = 33 # dodajemy element wiek (nie ma takiego klucza - więc zostanie dodany)
print(dane)
dane['wzrost'] = 180 # zmieniamy wartość (jest taki klucz- więc zmienia się wartość)
#lub
dane['wiek'] = dane['wiek'] + 1 # tak jak ze zwykłą zmienną
print(dane)
del dane['wiek'] #usuwamy element
print(dane)
```

```python
# chcemy mieć dostęp do kluczy / wartości
print(dane.keys()) # zwraca widok kluczy
print(list(dane.keys()))
print(list(dane.values()))
```

```python
# wartością słownika może być np. lista,
# kluczem może być liczba
faktury_kwota = {
    1: [1200,400,300], # zaleca się 4 spacje
    "marek": [200,300],
    "jarek": 200, # zaleca się dodanie przecinka, aby w przyszłości łatwiej edytować tekst
    }                   # (nic nie zmienia w działaniu programu)
print(faktury_kwota)
print(faktury_kwota[1])
print(type(faktury_kwota[1]))
print(faktury_kwota['marek'][1])
print(sum(faktury_kwota[1]))
```

```python
#metody
print(faktury_kwota.get("marek"))
print(faktury_kwota.get("zenek"))
print(faktury_kwota.get("zenek", "Brak danych"))
# print(faktury_kwota['zenek']) tak otrzymamy błąd
```

## Sets typ set - 6
Nieuporządkowany zestaw unikalnych obiektów <br>
{"arek", "jarek"}
<h1> { ... , ... }

```python
z = {"arek", "darek", "arek"} # indeksy nie działają
print(z)
type(z)
```

```python
z.add("marek") # zbiory są mutowalne - możemy np dodać element
print(z)
z.remove("arek") # lub usunąć
print(z)
```

```python
lista = ["arek", "darek", "arek"] # usuwanie duplikatów z listy
print(lista)
lista = list(set(lista))
lista
```

```python
# frozenset - zbiór niemutowalny - nie będzie można go zmienić
z = frozenset({"arek", "darek", "arek"})
print(z)
print(type(z))
z.remove("arek")
```

# Operatory

## Porównania 'Comparison operation'

```python
print(1 == 1)
print(2 != 3)
print(3 <= 3)
print(5 > 2)
```

## Przypisania 'Assigment Operators'

```python
x = 100
x += 3 # x = x + 3
print(x)
x *= 4 # x = x * 4
print(x)
x **= 2 # x = x ** 2
print(x)
```

## Logiczne 'Logical operators'

```python
a = 'arek'
b = 'marek'
print(a == b and 2 != 3) # komiunkcja jest prawdziwa tylko gdy oba jej czynniki są prawdziwe
print(a == b or 2 != 3) # alternatywa jest fałszywa tylko gdy oba czynniki są fałszywe
print(1 != 1 or False)
print(not(2 != 20)) # negacja
```

## Operatory arytmetyczne 'Arithmetic operators'

`+ - * / % modulus // floor devision ** exponent`

```python
print(7 % 3) # mamy 7 jabłek i chcemy je podzielić na 3 osoby (nie krojąc jabłka) - ile zostanie (reszta z dzielenia)
print(7 // 3) # ta sama sytuacja - ile dostanie każdy? (w praktyce - zaokrąglenie w dół) (dzielenie całkowite)
print(2 ** 3) #potęga
print(4 ** (1/2)) # pierwiastek 2 stopnia
print(4 ** 0.5) # pierwiastek 2 stopnia
print(4 ** .5) # pierwiastek 2 stopnia
# nawiasy okrągłę (dla ułatwienia są kolorowane)
2**8  == ((2+2)**(2+2))
```

## Operatory specjalne 'Special oparator'


### Operator identyczności 'Identity operators'
Sprawdza czy to jest ten sam obiekt (nie czy jest **taki sam** (==) tylko czy to **ten sam**) <br>
czy zmienne (wartości) zajmują tą samą część pamięci (operatory identyczności)

```python
a = 'arek'
b = 'marek'
c = 'arek'
print(a is b)
print(a is not b)
print(a is c) # ten sam obiekt
print(id(a))
print(id(c))
```

```python
a = [1,2,3]
b = a.copy()
print(a)
print(b)
print(a is b)
a == b
```

### Operator przynależności 'Membership operators'
sprawdza czy jest dany element w liście, zbiorze, słowniku, krotce, obiekcie str

```python
imiona = ['arek', ' marek', 'darek']
print('arek' in imiona)
print('darek' not in imiona)

osoba = {'arek':1, 'darek':2}
print('arek' in osoba) # sprawdza czy klucz jest częścią słownika
print(1 in osoba) # nie sprawdza wartości
# metoda values()
print(1 in osoba.values()) # tak sprawdzimy czy wartość jest częścią słownika

print('s' in 'socjologia')
```
