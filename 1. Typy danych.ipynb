### Nazwy zmiennych
**Identifier** - np nazwa funkcji, listy czy zmiennej
wszystko jest case-sensitive (wilekość liter ma znaczenie)
1. Samoopisujące - np. name
2. Bez spacji - może być np my_name (Snake Case) lub myName (Camel Case)
3. Nie stosujemy symboli takich jak  !, @, #, $ itd
4. Nie zaczynamy od liczby
5. Nie używamy słów które mają szczególne znaczenie dla pythona

### Obiekty mutowalne i niemutowalne
Każdy typ danych - lista, zmienna int, float słownik to obiekt

```python
# Niemutowalny
# tworzymy obiekt w którym jest 1 - nazywamy go a
a = 1
print(id(a))  # sprawdzamy miejsce w pamięci
# tworzymy drugą etykietę b i podpinamy ją do tego samego obiektu
b = a
print(id(b)) # sprawdzamy miejsce w pamięci
# przypisujemy do a 2 co oznacza że a nie wkazuje już na obiekt w którym jest 1 tylko na nowy w którym jest 2
a = 2
# w pierwszym obiekcie nadal jest 1
print(id(a))
#gdzie jest zarezerwowana pamięć na moja zmienną - tworzenie zmiennej = przydzielanie przestrzeni w pamięci komputera

print('a jest ', a, 'b jest ', b)
```

## Dynamiczne typowanie (Dynamic typing)
Mogę raz przypisać int do danej zmiennej, a później do tej samej zmiennej przypisać float (zmienić jej typ)

```python
number = 10
print(type(number))

number = "arek"
print(type(number))
```

```python
user_number = input('Enter a number ') #funcja input
print('Your number is: ', user_number) #funkcja print
type(user_number)
```

```python
number1 = int(input('podaj ilosc: ')) #konwertowanie stringa na int
number2 = int(input('podaj cene: '))
wartosc = number1 * number2
print('wartos wynosi: ', wartosc)
podatek = 0.23
wartosc_podatku = wartosc * podatek
wartosc_podatku
```
