## Dynamiczne typowanie (Dynamic typing)
Mogę raz przypisać **int** do danej zmiennej, a później do tej samej zmiennej przypisać **str** (zmienić jej typ) <br>
Nie deklaruję wcześniej jaka to będzie zmienna

```python
number = 10
print(type(number))

number = "arek"
print(type(number))
```

```python
user_number = input('Enter a number ')
# funcja input wyświetli "Enter a number" - użytkownik ma coś wpisać z klawiatury i wcisnąć Enter. Wtedy:
# zwróci str (to co wpisze użytkownik)
# obiekt str zostanie przypisany do zmiennej user_number
print('Your number is: ', user_number) #funkcja print
type(user_number)
```

```python
number1 = int(input('podaj ilosc: ')) #konwertowanie stringa na int za pomocą finkcji int()
number2 = int(input('podaj cene: '))
wartosc = number1 * number2
print('wartos wynosi: ', wartosc)
podatek = 0.23       # można też zapisać jako .23
wartosc_podatku = wartosc * podatek
wartosc_podatku
```
