##Lambda function
funkcja anonimowe bez nazwy

```python
def dodaj10(a):
  return a+10

print(dodaj10(2))


x = lambda a: a+10    # zapisujemy w jednej linijce

print(x(2)) # tu w nawiasie zwracany jest wynik
print(x)
y = x(5)    # to co podamy tu jako argument przeniesie do lambdy
print(y)

# Przy przypisywaniu nie widaÄ‡ tu wiÄ™kszych korzyÅ›ci
# moÅ¼emy teÅ¼ (w taki sposÃ³b) stosowaÄ‡ je wielokrotnie
```

```python
round((lambda m,n: m / n)(1,3), 10) # lambdÄ™ moÅ¼emy umieÅ›ciÄ‡ w funkcji - tu round
                                    # teraz widzimy, Å¼e korzystamy z niej tylko raz
```

```python
(lambda x: x.capitalize())('arek')
```

```python
# Funkcje wyÅ¼szego rzÄ™du

def powitanie(lam, imie): # powitanie - funkcja wyÅ¼szego rzÄ™du
  imie = lam(imie)
  print(f"Witaj {imie}")

powitanie(lambda x: x.capitalize(), 'arek') # przyjmuje jako argument drugÄ… funkcjÄ™
```

##Scope (zakres)
*Obszar w ktÃ³rym mamy dostÄ™p do danej zmiennej* <br>
**Global Variable**
poza funkcjÄ… - moÅ¼na ich uÅ¼yÄ‡ wszÄ™dzie (np. w Å›rodku funkcji) <br>
**Local Variable**
moÅ¼e byÄ‡ uÅ¼yta tylko w funkcji


```python
zmienna = 'zewnÄ…trz'

def drukuj():
    zmienna = 'wewnÄ…trz'
    return zmienna

print(drukuj())
print(zmienna)
```

**ReguÅ‚a LEGB** <br>
Local -> Enclosing (funkcja nadrzÄ™dna)-> Global -> Built in <br>
Szukanie zmiennej w rÃ³Å¼nych zakresach

```python
liczba = 10

def zew():
  def wew():
    print(liczba**2) # print - przykÅ‚ad poziomu "Built in" wbudowanego
  wew()
zew()
```
