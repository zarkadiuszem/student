## Funkcje zdefiniowane przez użytkownika

```python
def drukowanie(): # słowo kluczowe def nawa_funkcji():
    print('Hej')

drukowanie()


print(2+3)  # tę funkcję możemy wywołać bez definiowania bo jest wbudowana
drukowanie() # wywołujemy funkcję bez argumentu drugi raz
```

```python
def dodawanie(a, b): # w nawiasie podane są parametry funkcji. Są to "a" i "b"
  """Funkcja oblicza sumę dwóch liczb"""    #docstring - informacja o tym co funkcja robi
  suma = a + b
  return suma

c = dodawanie(2, 3)  # w wywołaniu funkcji podajemy dwa argumewnty 2 i 3 (argumenty przekazywane są do treści funkcji)
                     # mamy tu argumenty pozycyjne
print(f'"c" jest teraz {c}')
```

```python
# argumenty pozycyjne
def odejmowanie(a,b):
  pass                # jeśli funkcja (na tym etapie) nie ma nic robić wpisujemy pass (słowo kluczowe)
odejmowanie(1,2)
```

```python
# argumenty domyślne
def dodawanie(a, b=10):  # domyślnie za b wstawiamy 10 jeśli nie jest ten argument podany
                         # b = 10 musi być na drugim miejscu, żeby nadal działały argumenty pozycyjne
                         # czyli - domyślne na koniec
    razem = a+b
    print(razem)
suma = dodawanie(2)    # nic nie zwróciliśmy - więc nie ma czego przypisać
print(suma)
```

Gdy nie wiadomo ile będzie argumentów (arbitrary argument)

```python
def osoby(*imiona):   # * to unpacking operator (operator rozpakowania)
  print(type(imiona))   # uzyskujemy krotkę
  print(imiona)
  for imie in imiona:
    print('Czołem ', imie.capitalize())

osoby('jarek', 'darek', 'marek')
```

```python
lista = [['arek', 'jarek'], ['darek','arek']] # listy w liście

def pracownicy(lista):
  print(lista)
  for imiona in lista:
    print('Czołem ', imiona)

pracownicy(lista)
```

```python
lista = [['arek', 'jarek'], ['darek','arek']]

def pracownicy(lista):
  for imiona in lista:  # zagnieżdżona pętla
    for imie in imiona:
      print('Czołem ', imie.capitalize())

pracownicy(lista)
```

##Lambda function
funkcja anonimowe bez nazwy

```python
def dodaj10(a):
  return a+10

print(dodaj10(2))


x = lambda a: a+10    # zapisujemy w jednej linijce

print(x(2)) # tu w nawiasie zwracany jest wynik
print(x)
y = x(5)    # to co podamy tu jako argument przeniesie do lambdy
print(y)

# Przy przypisywaniu nie widać tu większych korzyści
# możemy też (w taki sposób) stosować je wielokrotnie
```

```python
round((lambda m,n: m / n)(1,3), 10) # lambdę możemy umieścić w funkcji - tu round
                                    # teraz widzimy, że korzystamy z niej tylko raz
```

```python
(lambda x: x.capitalize())('arek')
```

```python
# Funkcje wyższego rzędu

def powitanie(lam, imie): # powitanie - funkcja wyższego rzędu
  imie = lam(imie)
  print(f"Witaj {imie}")

powitanie(lambda x: x.capitalize(), 'arek') # przyjmuje jako argument drugą funkcję
```

##Scope (zakres)
*Obszar w którym mamy dostęp do danej zmiennej* <br>
**Global Variable**
poza funkcją - można ich użyć wszędzie (np. w środku funkcji) <br>
**Local Variable**
może być użyta tylko w funkcji


```python
zmienna = 'zewnątrz'

def drukuj():
    zmienna = 'wewnątrz'
    return zmienna

print(drukuj())
print(zmienna)
```

**Reguła LEGB** <br>
Local -> Enclosing (funkcja nadrzędna)-> Global -> Built in <br>
Szukanie zmiennej w różnych zakresach

```python
liczba = 10

def zew():
  def wew():
    print(liczba**2) # print - przykład poziomu "Built in" wbudowanego
  wew()
zew()
```

## Rekursja / rekurencja
Funkcja wywołuje samą siebie

```python
def inkrementacja(x):
    print(f"Głębokość rekursji: {x}")
    if x == 5:
        return   # kończy działanie funkcji
    else:
        inkrementacja(x + 1)

inkrementacja(0)
```
